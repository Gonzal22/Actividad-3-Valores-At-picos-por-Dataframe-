```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel("Gastos y costos 20-23.xlsx",header=3,sheet_name="2023")
df
```

```python
df.info()
```

reemplazar valores nulos

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
df =df.fillna(method="bfill")
df
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
df['MP'] =df['MP'].fillna(method="ffill")
df['FP'] =df['FP'].fillna(method="ffill")
df['Poliza'] =df['Poliza'].fillna(method="ffill")
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

valores atípicos

```python
fig = plt.figure(figsize =(15, 8))
df.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

```python
y=df
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
data2= df[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data2
```

```python
data3= data2.drop(['Fecha',"Folio","UUID","RFC","Proveedor",
                   "TIPO GASTO","Descripción","MP","Tipo","Status","Poliza","Otros"], axis=1)
data3
```

```python
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
fig = plt.figure(figsize =(15, 8))
data_clean.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
y=data3

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
data3_iqr= data3[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data4_iqr.mean(),1))
data4_iqr
```

```python
fig = plt.figure(figsize =(15, 8))
data4_iqr.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```
